```python
import pandas as pd
df = pd.read_csv('dataset_komentar_youtube.csv')
df.head()
```

# **Cleaning & Preprocesing**

```python
import re
def clean_text(text):
    # Hapus URL
    text = re.sub(r'http\S+', '', text)
    # Hapus karakter non-alphabet
    text = re.sub(r'[^A-Za-z\s]+', '', text)
    # Ubah ke huruf kecil
    text = text.lower()
    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Terapkan fungsi pembersihan ke kolom 'textDisplay'
df['TextDisplay'] = df['textDisplay'].apply(clean_text)

# Lihat hasil dari 5 data pertama setelah cleaning
df[['textDisplay', 'TextDisplay']].head()
```

```python
def clean_author(author):
    # Hapus simbol atau karakter khusus
    return re.sub(r'[^A-Za-z0-9\s]+', '', author)

df['author'] = df['authorDisplayName'].apply(clean_author)

# Lihat hasil dari kolom 'authorDisplayName' yang telah dibersihkan
df[['authorDisplayName', 'author']].head()
```

```python
# Ubah kolom 'publishedAt' menjadi format datetime
df['publishedAt'] = pd.to_datetime(df['publishedAt'])

# Ubah kolom 'likeCount' menjadi integer
df['likeCount'] = pd.to_numeric(df['likeCount'], errors='coerce')
```

```python
# Ubah kolom 'likeCount' menjadi tipe numerik
df['likeCount'] = pd.to_numeric(df['likeCount'], errors='coerce')

# Mengisi nilai NaN di 'likeCount' dengan 0 jika ada nilai kosong
df['likeCount'].fillna(0, inplace=True)
```

```python
# Hapus kolom yang di inginkan
df.drop(columns=['author'], inplace=True)

# Tampilkan DataFrame setelah kolom dihapus
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Contoh tokenisasi menggunakan CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['TextDisplay'])

# Hasil tokenisasi dalam bentuk matriks
print(X.toarray())
```

```python
from sklearn.preprocessing import StandardScaler

# Scaling pada kolom 'likeCount'
scaler = StandardScaler()
df['likeCount_scaled'] = scaler.fit_transform(df[['likeCount']])

# Tampilkan hasil scaling
df[['likeCount', 'likeCount_scaled']].head()
```

```python
df.drop(columns=['likeCount'], inplace=True)

# Tampilkan DataFrame setelah kolom dihapus
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Membuat TF-IDF vectorizer
tfidf = TfidfVectorizer(max_features=1000)

# Representasi teks menggunakan TF-IDF
X_tfidf = tfidf.fit_transform(df['TextDisplay']).toarray()
df.head()
```
